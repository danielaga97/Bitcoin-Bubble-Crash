## Caracter√≠sticas del modelo

Siguiendo a Cheah y Fry (2015) se asume que el precio de Bitcoin ùëÉ(ùë°) describe un proceso estoc√°stico:

$ùëÉ(ùë°) = ùëÉ_1(ùë°)(1 ‚àí ùúÖ)ùëó(ùë°)$ [1]

Donde $ùëÉ_1(ùë°)$ satisface:
$ùëëùëÉ_1(ùë°) = [ùúá(ùë°) + ùúé^2(ùë°)/2]ùëÉ_1(ùë°)ùëëùë° + ùúé(ùë°)ùëÉ_1(ùë°)ùëëùëä_ùë°$ [2]

Siendo $ùëä_ùë°$ un proceso de Weiner y ùëó(ùë°) un proceso de salto que implica:

$ùëó(ùë°) = { 0 antes del crash; 1 despues√©s del crash $ [3]

As√≠, cuando un crash ocurra, ùúÖ% es autom√°ticamente eliminado del valor del activo. Antes del crash $ùëÉ(ùë°) = ùëÉ_1(ùë°)$, se define $ùëã_ùë° = log(ùëÉ(ùë°))$ que satisface:

$ùëëùëã_ùë° = ùúá(ùë°)ùëë_ùë° + ùúé(ùë°)ùëëùëä_ùë° ‚àí ùë£ùëëùëó(ùë°)$ [4] 

Aqu√≠, el estad√≠stico de prueba ùë£ = $‚àí ln[(1 ‚àí ùúÖ)] > 0$ muestra como la burbuja impactar√° sobre los precios observados. Suponiendo que el crash a√∫n no ha ocurrido para el tiempo ùë°, en este caso se tiene:

$ùê∏[ùëó(ùë° + ‚àÜ) ‚àí ùëó(ùë°)] = ‚àÜ‚Ñé(ùë°) + ùëú(‚àÜ)$, [5]
$ùëâùëéùëü[ùëó(ùë° + ‚àÜ) ‚àí ùëó(ùë°)] = ‚àÜ‚Ñé(ùë°) + ùëú(‚àÜ)$, [6]

Donde ‚Ñé(ùë°) es la tasa de riesgo.


La estimaci√≥n se construir√° con base en dos supuestos:

‚ñ™ Supuesto 1: La tasa intr√≠nseca de retorno es asumida constante e igual a ùúá (deriva).

$ùúá(ùë°) = ùúá + ùë£‚Ñé(ùë°)$ [7]

‚ñ™ Supuesto 2: El nivel intr√≠nseco de riesgo es asumido constante e igual a $ùúé^2$ (volatilidad).

$ùúé^2(ùë°) = ùúé^2 ‚àí ùë£^2‚Ñé(ùë°)$  [8]

La interacci√≥n entre riesgo y retorno gu√≠a el comportamiento de mercados financieros. Finalmente, la formaci√≥n una burbuja especulativa se determina contrastando un par√°metro llamado ùíó ‚Äìpropio de la modelaci√≥n estoc√°stica‚Äì, el cual mide el riesgo de un crash ‚Äìca√≠da abrupta del precio‚Äì. 

$ùêª_ùëú: ùë£ = 0$ No hay una burbuja especulativa en formaci√≥n.
$ùêª_ùëú: ùë£ > 0$ Existe una burbuja especulativa en formaci√≥n.

Si ùë£ = 0 el precio fundamental es:
$ùëÉ_ùëì(ùë°): = ùê∏(ùëÉ(ùë°)) = ùëÉ(0)ùëí^(ùúá~ùë°)$ donde $ùúá~= ùúá + ùúé^2/2$ [9]

Para estimar el estad√≠stico ùíó se usa la metodolog√≠a conocida como **Ley de Potencias Log-Peri√≥dica LPLL** (Johansen, Ledoit, & Sornette, 2008). 
La LPLL estima unos par√°metros que se relacionan con el proceso estoc√°stico, seg√∫n el siguiente sistema de ecuaciones no lineal:
$ùê∂ = ‚àíùúÖùõºùõΩ/‚àöùõΩ^2 + ùúî^2$ Equivalencia para C [10]
$ùõΩ = 1 ‚àí ùõº$ Equivalencia para ùõΩ [11]
$ùêµ = ‚àíùúÖùõº/ùõΩ$ Relaci√≥n con el porcentaje de ca√≠da ùúÖ [12]
$ùúá(ùë°) = ùúÖ‚Ñé(ùë°)$ Equivalencia para ùúá(ùë°) [13]
$‚Ñé(ùë°) ‚âà ùêµ‚Ä≤(ùë°_ùëê ‚àí ùë°)‚àíùõº$ Relaci√≥n con tasa de riesgo ‚Ñé(ùë°) [14]

El movimiento browniano es 

```python
import numpy as np
import matplotlib.pyplot as plt
N = len(Precios)      #intervalos
T = 150        #tiempo de simulacion
dt =T/N      #ancho del intervalo
tray = 100     #numero de trayectorias a simular
mu = 0.15        #deriva
sigma = .3    #volatilidad
S_0 = 20       #valor inicial



S=np.empty((tray,N))
S[:,0]=S_0
for i in range(N-1):
    S[:,i+1]=S[:,i]*np.exp((mu-0.5*sigma**2)*dt + np.random.randn(tray)*np.sqrt(dt)*sigma)

plt.title('Geom Brownian Motion', fontsize=22)
plt.xlabel('time', fontsize=12)
plt.ylabel('S_t', fontsize=12)
plt.grid(1)
plt.plot(S.T)
plt.show()
```

Se mostrar√° una versi√≥n vectorizada que grafica 10 trayectorias del
 proceso de *wiener*
Cumsum es la suma acumulativa, la cual va sumando los incrementos normales  $(0,d_t)$
Cumsum por si s√≥lo no nos permite la condicion $W_0 = 0$, entonces *tray son los ceros que les hace falta a las trayectorias para que inicien en 0 y se los agregamos con el comando append $axis = 1$ nos asegura que se agregen por columna
 ($axis = 0$ es por fila)

```python
import matplotlib.pyplot as plt
N = len(Precios)   #intervalos
T= 150     #tiempo de simulacion
dt=T/N  #ancho del intervalo
tray= 10 #numero de trayectorias a simular
W = np.cumsum(np.sqrt(dt) * np.random.randn(tray,N),axis=1) 
W = np.append([[0]]*tray,W, axis = 1)       #wienner process


plt.title('Brownian Motion- Wienner', fontsize=22)
plt.xlabel('time', fontsize=12)
plt.ylabel('W_t', fontsize=12)
plt.grid(1)
plt.plot(W.T)
plt.show()
```

## Martingalas

Recordemos que una secuencia {Xn}n de variables aleatorias con valores en un espacio finito ‚Ñ¶ es una martingala cuando
$E(X_(n+1)|X_1 = x_1, . . . , X_n = xn) = x_n$,
siendo $x_i$ se denomina estado del sistema en el instante i.
Se dice que es una supermartingala cuando
$E(X_(n+1)|X_1 = x_1, . . . , X_n = x_n) ‚â§ x_n$,
y una submartingala cuando
$E(X_(n+1)|X_1 = x_1, . . . , X_n = x_n) ‚â• x_n$
En esta pr√°ctica, vamos a considerar dos ejemplos de martingalas y vamos a usar simulaci√≥n
para estimar dos probabilidades asociadas a estos procesos
